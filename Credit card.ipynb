```python
credit_card_numeric_features = credit_card_new.select_dtypes(include=[np.number])
credit_card_num =credit_card_numeric_features.drop('TENURE',axis=1)
credit_card_num.head()
```

```python
credit_card_num.shape
```

```python
from sklearn.preprocessing import StandardScaler
credit_card_num_std = StandardScaler().fit_transform(credit_card_num)
print(credit_card_num_std)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=11, random_state=0)
PrincipalComponents = pca.fit_transform(credit_card_num_std)
```

```python
PCA_df = pd.DataFrame(data = PrincipalComponents, columns = ['PC1', 'PC2', 'PC3','PC4','PC5', 'PC6', 'PC7', 'PC8','PC9','PC10','PC11'])
```

```python
PCA_df.head()
```

```python
var = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=3)*100)
print(var)
```

```python
cov_mat = np.cov(credit_card_num_std.T)
```

```python
print(cov_mat[0:5])
```

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)
```

```python
print('Eigenvalues:','\n','\n', eig_val,"\n")

print('Eigenvectors:','\n','\n',eig_vec,'\n')
 
```

```python
eig_val = list(eig_val)
eig_val.sort(reverse = True) 
print(eig_val)
```

```python
plt.plot(eig_val) 
plt.xlabel('Principal Components') 
plt.ylabel('Percentage of explained variance')
plt.annotate('Elbow Point', xy=(3,1.25), xytext=(5, 2.5), arrowprops=dict(facecolor='black', arrowstyle = 'simple'))
plt.title('Scree Plot')
plt.show()  
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

n_clusters = [2,3,4, 5, 6,7, 8, 9,10]


for K in n_clusters :
    cluster = KMeans(n_clusters= K, random_state= 10)
    predict = cluster.fit_predict(credit_card_num_std)
    
    score = silhouette_score(credit_card_num_std, predict, random_state= 10)
    print ("For n_clusters = {}, silhouette score is {})".format(K, score))
```

```python
model = KMeans(n_clusters=5, random_state= 10)
model.fit(credit_card_num_std)
```
